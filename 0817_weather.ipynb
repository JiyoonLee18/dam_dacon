```python
# 2022년 데이터 가지고 오기
weather_2022=pd.DataFrame()
for i in range(1, 8):
    if i==2:
        j=28
    elif i==4 or i==6 :
        j=30
    else:
        j=31
    i='0'+str(i)
    j=str(j)
    start_month='2022'+i+'01'
    end_month='2022'+i+j
    params = f'?{quote_plus("ServiceKey")}={enKey}&' + urlencode({ 
        quote_plus("pageNo"): "1", # 페이지 번호 // default : 1 
        quote_plus("numOfRows"): "999", # 한 페이지 결과 수 // default : 10 
        quote_plus("dataType"): "JSON", # 응답자료형식 : XML, JSON 
        quote_plus("dataCd"): "ASOS",
        quote_plus("dateCd"): "HR",
        quote_plus("startDt"):start_month, 
        quote_plus("startHh"): "00",
        quote_plus("endDt"): end_month,
        quote_plus("endHh"): "23",
        quote_plus("stnIds"): "108"
        })

    # URL parsing 
    req = urllib.request.Request(url + params)

    # Get Data from API 
    response_body = urlopen(req).read()   # get bytes data 
    response_body
    # # Convert bytes to json 
    data = json.loads(response_body) 
    data
    # Result 
    res = pd.DataFrame(data['response']['body']['items']['item'])
    weather_2022=pd.concat([weather_2022, res])
    time.sleep(2)
```

```python
def make_datetime(df):
    df['date']=pd.to_datetime(df['tm'])
    df['year']=df['date'].dt.year
    df['month']=df['date'].dt.month
    df['day']=df['date'].dt.day
    df['day'] = df['date'].dt.day
    df['dayofweek'] = df['date'].dt.dayofweek
    df['hour']=df['date'].dt.hour
    return df
```

```python
weather_df=make_datetime(weather_df)
weather_2022=make_datetime(weather_2022)
```

```python
# 2012년 11월 29일 2시데이터 없어서 1시 데이터 복제해서 넣어줌
tmp=weather_df.copy()
tmp_2012=tmp[tmp['year']==2012]
tmp=pd.DataFrame(tmp_2012[tmp_2012['month']==11][tmp_2012[tmp_2012['month']==11]['day']==29].iloc[1]).T
```

```python
weather_df=pd.concat([weather_df, weather_2022, tmp])
weather_df.sort_values(by='tm', inplace=True)
weather_df.reset_index(inplace=True, drop=True)
```
